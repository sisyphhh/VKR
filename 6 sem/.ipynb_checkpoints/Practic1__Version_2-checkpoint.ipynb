```python
#print("Allel`s ficsation in 0-10 sysytem: 0 or 10 \"death\" ficsation")
import numpy as np
from matplotlib import pylab as plt
from matplotlib import rcParams
%matplotlib inline

import random
import os
import pandas as pd

from scipy.stats import gamma
from scipy.stats import erlang
```

```python
#необходимые переменные: гянуть библиотеку Pandas
eps=0.00001
NNN=5

liamda_i_massive=[]
matrix_liambda=[]

for i in range(2*NNN +1):
    liamda_i_massive.append(round((2*NNN-i)*(i/(2*NNN)),5))

print('liamda_i_massive: ',liamda_i_massive)

for i in range(2*NNN +1):
    matrix_liambda.append([])
    #print(matrix_liambda)
    for j in range(2*NNN +1):
        if (i == 0) or (i == 10):
            matrix_liambda[i].append(0.0)
        else:
            if j==i:
                matrix_liambda[i].append(round((-2)*liamda_i_massive[i],5))
            elif (j==i-1) or (j==i+1):    
                matrix_liambda[i].append(round(liamda_i_massive[i],5))
            else:
                matrix_liambda[i].append(0.0)

print('\nLiambda matrix A:')
for i in range(2*NNN +1):
    print()
    for j in range(2*NNN +1):
        print(matrix_liambda[i][j],' ',end='')
```

```python
def F_calc(xxx,aa,sscale,Bord_a=0,Bord_b=0): 
    F=round(gamma.cdf(xxx,a=aa,scale=sscale),6)
    #print(F)
    #x=[0,-> ~ 10] if scale =1/abs(liambda) 
    #a=k =2 or 3
    #scale = 1/abs(liambda) so it`s 1/|mart[i][i]|
    if Bord_a!=0:
        return round(F - Bord_a,6)
    elif Bord_b!=0:
        return round(F - Bord_b,6)
    else:
        return F

def dihotomy(x,aa=2,sscale=1/1,eps=0.01,bord='A',bordsc=0):#x=[a,b] a<b ,A_bord=0,B_bord=0
    x_1=[]
    for i in x:
        x_1.append(i)
    #tmp=[]
    counter=0
    while abs(x_1[1]-x_1[0])>eps:
            x_finalle=(x_1[0]+x_1[1])/2
            #print(counter,x_finalle,x_1)
            counter+=1
            
            if bord=='A':
                if F_calc(x_1[0],aa,sscale,Bord_a=bordsc)*F_calc(x_finalle,aa,sscale,Bord_a=bordsc)<0:
                    x_1[0]=x_1[0]
                    x_1[1]=x_finalle

                x_finalle=(x_1[0]+x_1[1])/2

                if F_calc(x_1[0],aa,sscale,Bord_a=bordsc)*F_calc(x_finalle,aa,sscale,Bord_a=bordsc)>0:
                    x_1[0]=x_finalle
                    x_1[1]=x_1[1]    
            else:#bord=='B':
                if F_calc(x_1[0],aa,sscale,Bord_b=bordsc)*F_calc(x_finalle,aa,sscale,Bord_b=bordsc)<0:
                    x_1[0]=x_1[0]
                    x_1[1]=x_finalle

                x_finalle=(x_1[0]+x_1[1])/2

                if F_calc(x_1[0],aa,sscale,Bord_b=bordsc)*F_calc(x_finalle,aa,sscale,Bord_b=bordsc)>0:
                    x_1[0]=x_finalle
                    x_1[1]=x_1[1]
            #print(x_1)
    #print(x_finalle)
    '''if bord=='A':
        print('in this dot: ',F_calc(x_finalle,aa,sscale,Bord_a=bordsc))
    else:
        print('in this dot: ',F_calc(x_finalle,aa,sscale,Bord_b=bordsc))'''
    return x_finalle
```

```python
def liambda_disturb_generator(mass,start_i,disturb_n='expon',kkk=1):#генерация времени прибывания с  усечением
    Alfa1_test=0.2#усечение
    Alfa2_test=0.2#усечение может быть 0.2?
    Alfa2_test=1-Alfa2_test#для симметрии ограничений* 
    #liambda=abs(matr[start_i][start_i])#диагональный элемент
    liambda=abs(mass[start_i])
    
    if disturb_n=='expon':
        A_border=np.round(-np.log(1-Alfa1_test)/liambda,6)
        B_border=np.round(-np.log(1-Alfa2_test)/liambda,6)
        while True:
            time_generate=np.round(np.random.exponential(1/liambda),6)#depends of start_i : liambda=abs(matrix_to_r[start_i][start_i])
            if (A_border<=time_generate) and (B_border>=time_generate):
                break  
    else:#gamma 2k or 3k
        aas=kkk
        sscales=1/liambda
        op=sscales * aas * 2.5 #sscales * aas - mat ozhidanie
        A_border=dihotomy(x=[0,sscales * aas],aa=kkk,sscale=1/liambda,eps=0.01,bord='A',bordsc=0.2)
        B_border=dihotomy(x=[sscales * aas,op],aa=kkk,sscale=1/liambda,eps=0.01,bord='B',bordsc=0.8)
        while True:
            time_generate=np.asscalar(np.round(gamma.rvs(kkk,scale=1/liambda,size=1),6))
            if (A_border<=time_generate) and (B_border>=time_generate):
                break
        
    return time_generate
```

```python
def function_cycle(ii,liam_mass,test_num=100,disturb_name='expon',k=1,target=os.getcwd() + os.sep):
    
    with open (target+'table_%s_%s.txt'%(disturb_name,ii),'w') as outfile:
        outfile.write('start i '+'C()\t'+'t_left\t'+'\tt_right '+'\tNew i'+'\tTotal time'+' generation'+'\n') 
       
    t_total_end=0
    generation_total_end=0
    end_left_0,end_right_10=0,0
    i_final_end=0#to calcull where we stop 0 or 10
    
    for schetchik in range(test_num):
        generation_counter=0
        t_procces=0
        step_right=0
        step_left=0
        i=ii
        with open (target+'table_%s_%s.txt'%(disturb_name,ii),'a') as outfile:
            outfile.write('\n'+'test #'+str(schetchik)+'\n')
            
        while 1<=i<=9:
            j=i
            liambda=liam_mass[i]
            #print(liambda)
            step_left=liambda_disturb_generator(liam_mass,start_i=i,disturb_n=disturb_name,kkk=k)#round(np.random.exponential(1/liambda),5)
            step_right=liambda_disturb_generator(liam_mass,start_i=i,disturb_n=disturb_name,kkk=k)#round(np.random.exponential(1/liambda),5)
            if step_left<step_right:
                i-=1
                t_procces+=round(step_left,5)
            else:
                i+=1
                t_procces+=round(step_right,5)      
            generation_counter+=1
            #print(j,'\t','-- ','\t',step_left,'\t',step_right,'\t',i,'\t',round(t_total,5),'\t',generation_counter)
            with open (target+'table_%s_%s.txt'%(disturb_name,ii),'a') as outfile:
                outfile.write(str(j)+'\t--\t'+str(step_left)+'\t'+str(step_right)+'\t'+str(i)+'\t'+str(round(t_procces,5))+'\t'+str(generation_counter)+'\n')
        if i==0:
            end_left_0+=1
        else:
            end_right_10+=1
        generation_total_end+=generation_counter
        t_total_end+=t_procces
        i_final_end+=i
    
    t_total_end=round(t_total_end/test_num,5)
    generation_counter_end=round(generation_total_end/test_num)
    
    if round(i_final_end/(test_num*10))==0:
        i_final_end=0
    else:
        i_final_end=10
    
    end_left_0=round(end_left_0/test_num,5)
    end_right_10=round(end_right_10/test_num,5)
    
    #final answ for grafs
    listik=[ii,t_total_end,generation_counter_end,i_final_end,end_left_0,end_right_10]
    print('Our list is: ',listik)
    return listik
```

```python
import matplotlib.pyplot as plt
import pylab
import matplotlib.ticker

def lin_expon_graph(Y,f_size=(10,6),name='expon',labe='Time'):
    rcParams['figure.dpi']=100
    fig, ax = plt.subplots(figsize=f_size)
    main_target=catalogmaker(name='Graphics')
    x_i=[x for x in range(1,10)];print(x_i)
    
    plt.plot(x_i,Y,'v',label=labe +'_toFix')
    for i in range(len(Y)-1):
        plt.plot([x_i[i],x_i[i+1]],[Y[i],Y[i+1]],'--')
    
    plt.title('%s'%(name))
    plt.ylabel('%s'%(labe));plt.xlabel('Xi')
    plt.legend()
    plt.xticks(np.arange(1, 10, 1.0))
    
    if labe=='Time':
        plt.yticks(np.arange(min(Y)-1, max(Y)+2, 1))
    else:
        plt.yticks(np.arange(min(Y)-2, max(Y)+4, 2))   
    
    for i in range(len(Y)):
        if i<=len(Y)//2:
            ax.text(x_i[i]+0.2, Y[i], 'y=%s'%(round(Y[i],3)), fontsize=8)
        else:
            if labe=='Time':
                ax.text(x_i[i]-0.4, Y[i]-0.5, 'y=%s'%(round(Y[i],3)), fontsize=8)
            else:
                ax.text(x_i[i]-0.4, Y[i]-1, 'y=%s'%(round(Y[i],3)), fontsize=8)
                
    plt.grid(True)#plt.show()
    fig.savefig(main_target+name+labe)

    
def lin_graph(Y1,Y2,f_size=(10,6),name='gamma',code='N_T',labe='Time'):
    rcParams['figure.dpi']=100
    fig, ax = plt.subplots(figsize=f_size)
    main_target=catalogmaker(name='Graphics')  
    x_i=[x for x in range(1,10)];print(x_i)
    
    ax1 = fig.add_subplot(211)
    ax2 = fig.add_subplot(212)
    ax1 = plt.subplot2grid((2, 1), (0, 0))
    ax1.set_title(u'%s'%('Time'))
    ax2 = plt.subplot2grid((2, 1), (1, 0))
    ax2.set_title(u'%s'%('Generations'))
    
    
    ax1.scatter(x_i, Y1,marker='v', c='r', linewidth='1', edgecolor='black')
    for i in range(len(Y1)-1):
        ax1.plot([x_i[i],x_i[i+1]],[Y1[i],Y1[i+1]],linestyle='--',color='g' )#'orange'
    
    plt.yticks(np.arange(min(Y1)-1, max(Y1)+2, 1))
    for i in range(len(Y1)):
        if i<=len(Y1)//2:
            ax1.text(x_i[i]+0.2, Y1[i]+0.1, 'y=%s'%(round(Y1[i],3)), fontsize=8)
        else:
            ax1.text(x_i[i]-0.4, Y1[i]-0.5, 'y=%s'%(round(Y1[i],3)), fontsize=8)
    
    
    ax2.scatter(x_i, Y2, marker='v', c='r', linewidth='1',edgecolor='black')
    for i in range(len(Y2)-1):
        ax2.plot([x_i[i],x_i[i+1]],[Y2[i],Y2[i+1]],linestyle='--',color='g' )#'orange'
        
    plt.yticks(np.arange(min(Y2)-2, max(Y2)+4, 2)) 
    for i in range(len(Y2)):
        if i<=len(Y2)//2:
            ax2.text(x_i[i]+0.2, Y2[i]+0.1, 'y=%s'%(round(Y2[i],3)), fontsize=8)
        else:
            ax2.text(x_i[i]-0.4, Y2[i]-1, 'y=%s'%(round(Y2[i],3)), fontsize=8)
            
    # Создаем экземпляр класса, который будет отвечать за расположение меток
    locator2 = matplotlib.ticker.MultipleLocator (base=2)
    # Установим локатор для главных меток
    ax2.yaxis.set_major_locator (locator2)
        
    ax1.set_xlabel('$i$')
    ax1.set_ylabel('$ y1 $')
    ax2.set_xlabel('$i$')
    ax2.set_ylabel('$ y2 $')
    for ax in fig.axes:
        ax.grid(True)
    plt.tight_layout(h_pad = 1)
    fig.savefig(main_target+name+labe)
```
