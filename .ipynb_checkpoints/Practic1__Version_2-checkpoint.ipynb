## Allel`s ficsation in 9 sysytem: 0 or 10 "death" fixsation

```python
#print("Allel`s ficsation in 0-10 sysytem: 0 or 10 \"death\" ficsation")
import random
import numpy as np
from matplotlib import pylab as plt
from matplotlib import rcParams
%matplotlib inline
```

```python
#необходимые переменные: гянуть библиотеку Pandas
eps=0.00001
NNN=5

liamda_i_massive=[]
matrix_liambda=[]

for i in range(2*NNN +1):
    liamda_i_massive.append(round((2*NNN-i)*(i/(2*NNN)),5))

print('liamda_i_massive: ',liamda_i_massive)

for i in range(2*NNN +1):
    matrix_liambda.append([])
    #print(matrix_liambda)
    for j in range(2*NNN +1):
        if (i == 0) or (i == 10):
            matrix_liambda[i].append(0.0)
        else:
            if j==i:
                matrix_liambda[i].append(round((-2)*liamda_i_massive[i],5))
            elif (j==i-1) or (j==i+1):    
                matrix_liambda[i].append(round(liamda_i_massive[i],5))
            else:
                matrix_liambda[i].append(0.0)

print('\nLiambda matrix A:')
for i in range(2*NNN +1):
    print()
    for j in range(2*NNN +1):
        print(matrix_liambda[i][j],' ',end='')
```

```python
def function_cycle(ii):
    
    listik=[]#final answ for grafs
    listik.append(ii)
    
    round_mass=[]
    generation_counter=0
    t_total=0
    step_right=0
    step_left=0
    
    i_final_end=0#to calcull where we stop 0 or 10
    t_total_end=0
    generation_counter_end=0
    
    for schetchik in range(100):
        #generation_counter=0
        #t_total=0
        generation_counter=0
        t_total=0
        step_right=0
        step_left=0
        i=ii
        while 1<=i<=9:
            j=i
            liambda=liamda_i_massive[i]
            #print(liambda)
            step_left=round(np.random.exponential(1/liambda),5)
            step_right=round(np.random.exponential(1/liambda),5)

            if step_left<step_right:
                i-=1
                t_total+=round(step_left,5)
            else:
                i+=1
                t_total+=round(step_right,5)      

            generation_counter+=1
            print(j,'\t','-- ','\t',step_left,'\t',step_right,'\t',i,'\t',round(t_total,5),'\t',generation_counter)
        print()
        t_total_end+=t_total
        generation_counter_end+=generation_counter
        i_final_end+=i
    
    t_total_end=round(t_total_end/100,5)
    generation_counter_end=round(generation_counter_end/100)
    i_final_end=round(i_final_end/1000)
    
    listik.append(t_total_end)
    listik.append(generation_counter_end)
    listik.append(i_final_end)
    
    print('Our list is: ',listik)
    return listik
```

```python
#new2
#Math generations
print(liamda_i_massive,'\n')
#eps=0.00001
#NNN=5
#liamda_i_massive=[]
#matrix_liambda=[]

generation_counter=0
t_total=0
step_right=0
step_left=0

#i=int(input())# from 1...9
i=1
answer_list=[]
answer_list.clear()
print('start i ','C()\t','t_left\t','\tt_right ','\tNew i','\tTotal time',' generation')

for ii in range(2*NNN):
    
    #print('\n',ii,'\n')
    generation_counter=0
    t_total=0
    step_right=0
    step_left=0
    
    i=ii
    if 1<=i<=9:
        print('\nStart Position:',ii,'\n')
        b=function_cycle(i)
        answer_list.append(b)
        
    else:
        continue
```
